```python
import numpy as np
import matplotlib.pyplot as plt

# Grid
x = np.linspace(0, 1, 300)
y = np.linspace(0, 1, 300)
X, Y = np.meshgrid(x, y)

# Initialize Z
Z = np.zeros_like(X)

# Conditions
cond1 = (X < Y) & (Y < .5)
cond2 = (X < .5) & (Y > .5)
cond3 = (X > .5) & (X < Y)

# Apply piecewise function
Z[cond1] = 0.1 * (.5 - Y[cond1])
Z[cond2] = 0.9 * (Y[cond2] - .5)
Z[cond3] = 0.1 * (.5 - Y[cond3])

# Plot
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title('3D Plot of Piecewise Function')

fig.colorbar(surf, shrink=0.5, aspect=10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid
x = np.linspace(0, 10, 300)
y = np.linspace(0, 10, 300)
X, Y = np.meshgrid(x, y)

# Initialize
Z = np.zeros_like(X)

# Conditions
cond1 = (X < Y) & (Y < 5)
cond2 = (X < 5) & (Y > 5)

# Apply function
Z[cond1] = 0.1 * (5 - Y[cond1])
Z[cond2] = 0.9 * (Y[cond2] - 5)

# Plot
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')

# Labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title('3D Plot of Piecewise Function')

fig.colorbar(surf, shrink=0.5, aspect=10)

plt.show()
```

```python
lr = .1
delta = 0
T_test = 1000
k = alphas.shape[0]
Dalphas = np.diff([0] + list(alphas)+ [1])
B = 2

    
ds = np.zeros((T_test, k+1)) + B/(k+1)
    
    #qs = np.zeros((T_test, k))
qs      = np.cumsum(ds, 1)[:,::-1]
#qs[t+1] = np.cumsum(ds[t+1][::-1])[::-1]
covereds = np.zeros((T_test, k+1))
dcovereds = np.zeros((T_test, k+1))
grads = np.zeros((T_test, k+1))

scores = 2*np.random.beta(2,1, size=1000)#np.array(list(2*np.random.rand(1000)))

for t in tqdm(range(T_test)):
        #t_pred = t - ahead + 1
        t_pred = t
        covereds[t] = qs[t] >= scores[t]
        
        # Next, calculate the quantile update and saturation function
        #if t > 0:
        #    B = np.max(scores[:t])
        dcovereds[t] = np.logical_and(
                np.array(list(qs[t][1:])+[0]) < scores[t],
                qs[t] >= scores[t]
                )
        
        #Dgrad = np.array([-(1-dalpha) if dcovereds[t_pred, i] else dalpha 
        #                  for i, dalpha in enumerate(Dalphas)])
        
        Dgrad = np.array([-(1-dalpha) if dcovereds[t_pred, i] else dalpha 
                          for i, dalpha in enumerate(Dalphas)])
        
        grad = np.array([-alpha if covereds[t_pred, i] else (1-alpha) 
                         for i, alpha in enumerate(alphas)])
        grads[t] = Dgrad
        
        if t < T_test - 1:
            ds[t+1] = ds[t] * np.exp(lr * Dgrad)
            #ds[t+1,-1] = ds[t,-1] + lr * grad[-1]
            #ds[t+1] = ((1-delta)*ds[t+1]/np.sum(ds[t+1]) + delta/k)*B
            ds[t+1] = ds[t+1]/np.sum(ds[t+1])*B
            #ds[t+1] = np.clip(ds[t+1], .1, B*1.1)
            qs[t+1] = np.cumsum(ds[t+1][::-1])[::-1]
```

```python
import numpy as np
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.stats.diagnostic import het_arch

h = 12
# 1. Fit AR(p) on training window
model = AutoReg(inf_train, lags=3, old_names=False).fit()

# 2. In-sample diagnostics
print(model.summary())
resid = model.resid

# 3. Test for ARCH effects in residuals — the key diagnostic
arch_test = het_arch(resid, nlags=12)
print(f"ARCH LM p-value: {arch_test[1]:.4f}")  # reject → heteroscedastic

# 4. Forecast with prediction intervals
pred = model.get_prediction(start=len(inf_train), end=len(inf_train)+h-1)
ci = pred.conf_int(alpha=0.05)  # 95% intervals

# 5. Evaluate: coverage on held-out test set
actual = inf_test.values[:h]
covered = np.mean((actual >= ci["lower"]) & (actual <= ci["upper"]))
print(f"Empirical coverage: {covered:.2%}")  # compare to nominal 95%
```
