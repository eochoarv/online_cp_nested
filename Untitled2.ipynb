```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# ---------------------------
# Fake data to mimic the plot
# ---------------------------
np.random.seed(7)

colors = {
    "CP (PTIR)": "#d33682",          # magenta
    "CP (Raman)": "#0f9986",         # teal
    "MVCP (PTIR + Raman)": "#e0b700" # yellow
}

methods = list(colors.keys())

# x positions for the 2 big groups, with 3 methods inside each group
x_left  = [1.0, 1.4, 1.8]
x_right = [3.0, 3.4, 3.8]
x_all = x_left + x_right

# Top panel: empirical confidence means and error bars
acc_means = [0.950, 0.951, 0.949, 0.962, 0.971, 0.964]
acc_err   = [0.008, 0.020, 0.007, 0.004, 0.006, 0.005]

# Bottom panel: prediction set size samples
n = 120
size_data = {
    "left": {
        "CP (PTIR)": np.random.normal(1.08, 0.015, n),
        "CP (Raman)": np.random.normal(1.13, 0.020, n),
        "MVCP (PTIR + Raman)": np.random.normal(0.995, 0.015, n),
    },
    "right": {
        "CP (PTIR)": np.random.normal(1.90, 0.18, n),
        "CP (Raman)": np.random.normal(1.68, 0.14, n),
        "MVCP (PTIR + Raman)": np.random.normal(1.10, 0.03, n),
    }
}

# ---------------------------
# Figure layout
# ---------------------------
fig = plt.figure(figsize=(7.2, 6.2))
gs = fig.add_gridspec(2, 1, height_ratios=[1, 3], hspace=0.15)

ax_top = fig.add_subplot(gs[0])
ax_bot = fig.add_subplot(gs[1], sharex=ax_top)

# ---------------------------
# Top panel: error bars
# ---------------------------
for x, m, e, c in zip(x_all, acc_means, acc_err, list(colors.values()) * 2):
    ax_top.errorbar(
        x, m, yerr=e,
        fmt='o', ms=7,
        color=c, ecolor='black',
        elinewidth=1.8, capsize=4, capthick=1.8,
        zorder=3
    )

ax_top.axhline(0.95, color='gray', linestyle=(0, (10, 8)), linewidth=2)
ax_top.text(0.78, 0.905, "Target\nAccuracy", color='dimgray',
            fontsize=13, fontweight='bold', ha='left', va='center')

ax_top.set_ylim(0.85, 1.00)
ax_top.set_ylabel("Empirical\nConfidence", fontsize=16)
ax_top.set_yticks([0.85, 0.90, 0.95, 1.00])

# clean style
ax_top.spines['top'].set_visible(False)
ax_top.spines['right'].set_visible(False)
ax_top.spines['bottom'].set_visible(False)
ax_top.tick_params(axis='x', bottom=False, labelbottom=False)
ax_top.tick_params(axis='y', labelsize=12, width=1.8, length=8)
ax_top.spines['left'].set_linewidth(1.8)

# ---------------------------
# Bottom panel: boxplots + jitter
# ---------------------------
def add_box_and_points(ax, data, positions, color_order):
    for pos, (label, vals), color in zip(positions, data.items(), color_order):
        # jittered scatter
        jitter = np.random.uniform(-0.06, 0.06, size=len(vals))
        ax.scatter(
            np.full(len(vals), pos) + jitter, vals,
            s=20, facecolors='none', edgecolors=color,
            linewidths=1.2, alpha=0.8, zorder=1
        )

        # boxplot
        ax.boxplot(
            vals,
            positions=[pos],
            widths=0.22,
            patch_artist=False,
            showfliers=False,
            boxprops=dict(color='0.2', linewidth=2.5),
            medianprops=dict(color='0.2', linewidth=2.5),
            whiskerprops=dict(color='0.2', linewidth=2.5),
            capprops=dict(color='0.2', linewidth=2.5),
            zorder=2
        )

add_box_and_points(ax_bot, size_data["left"], x_left, list(colors.values()))
add_box_and_points(ax_bot, size_data["right"], x_right, list(colors.values()))

ax_bot.axhline(1.0, color='gray', linestyle=(0, (10, 8)), linewidth=2)
ax_bot.text(0.65, 0.93, "Target Set Size", color='dimgray',
            fontsize=13, fontweight='bold', ha='left', va='center')

ax_bot.set_ylabel("Average Prediction Set Size", fontsize=16)
ax_bot.set_ylim(0.8, 2.5)
ax_bot.set_yticks([1.0, 1.5, 2.0, 2.5])

ax_bot.set_xticks([1.4, 3.4])
ax_bot.set_xticklabels([
    "Spectral Dataset #1:\nUnchanged Subset",
    "Spectral Dataset #1:\nNoise-Added Subsets"
], fontsize=14)

ax_bot.spines['top'].set_visible(False)
ax_bot.spines['right'].set_visible(False)
ax_bot.tick_params(axis='both', labelsize=12, width=1.8, length=8)
ax_bot.spines['left'].set_linewidth(1.8)
ax_bot.spines['bottom'].set_linewidth(1.8)

# ---------------------------
# Legend
# ---------------------------
handles = [Patch(facecolor=colors[m], edgecolor=colors[m], label=m) for m in methods]
ax_bot.legend(
    handles=handles, loc='upper left', frameon=False,
    fontsize=14, handlelength=1.2, handletextpad=0.5
)

# x limits
ax_bot.set_xlim(0.6, 4.2)

plt.show()
```
